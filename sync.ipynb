```python
from dotenv import load_dotenv
import os

load_dotenv()

WORLD_BANK_DATASET_ID = os.getenv("WORLD_BANK_DATASET_ID")
AGRO_GOX_DATASET_ID = os.getenv("AGRO_GOX_DATASET_ID")
```

```python
DOMAIN = os.getenv("DOMAIN")
WORKSPACE_ID = os.getenv("WORKSPACE_ID")
API_KEY = os.getenv("API_KEY")
```

```python
import http.client

conn = http.client.HTTPSConnection(DOMAIN)

headers = {"API_KEY": API_KEY}
DATASET_ID = WORLD_BANK_DATASET_ID

conn.request(
    "GET",
    f"/api/v1/workspaces/{WORKSPACE_ID}/datasets/{DATASET_ID}/full",
    headers=headers,
)

res = conn.getresponse()
data = res.read()
```

```python
import json
import pandas as pd

# Parse the JSON data
parsed_data = json.loads(data.decode("utf-8"))

# Extract rows from the data
rows = parsed_data["data"]

# Create list of dictionaries with flattened structure
flattened_data = []
for row in rows:
    # Get the cells dictionary
    cells = row["cells"]
    # Add row_id to the cells dictionary
    cells["row_id"] = row["row_id"]
    flattened_data.append(cells)
```

```python
# Convert to DataFrame
df = pd.DataFrame(flattened_data)

# Replace 'â€¦' with NaN
df = df.replace("â€¦", pd.NA)

# Convert columns to numeric where possible
df = df.apply(pd.to_numeric, errors="ignore")

# Display first few rows
df.head()
```
